```python
import pandas as pd
```

## cls_attr 컴럼의 오브젝트 숫자로 바꿔주기

```python
trip['cls_attr'].unique()
```

```python
trip['cls_attr'] = trip['cls_attr'].replace("ui_bubble_rating bubble_10", 1.0)
trip['cls_attr'] = trip['cls_attr'].replace("ui_bubble_rating bubble_20", 2.0)
trip['cls_attr'] = trip['cls_attr'].replace("ui_bubble_rating bubble_30", 3.0)
trip['cls_attr'] = trip['cls_attr'].replace("ui_bubble_rating bubble_40", 4.0)
trip['cls_attr'] = trip['cls_attr'].replace("ui_bubble_rating bubble_50", 5.0)
```

```python
trip['cls_attr'].unique()
```

## 	Unnamed: 0 컬럼 삭제

```python
trip.drop(["Unnamed: 0"], axis=1)
```

## user_name 컬럼에 \n 뒤의 문자열 모두 제거

```python
trip['name'] = trip.user_name.str.split("\n").str[0]
```

```python
trip
```

```python
trip = trip.drop(["user_name"], axis=1)
# trip = trip.drop(["Unnamed: 0"], axis=1)
```

```python
trip = trip.drop(["food_name"], axis=1)
```

## 컬럼 순서 & 이름 정렬

- user_name, res_name, rating, rev_title, review, menu

```python
trip.columns
```

```python
trip.head()
```

```python
trip = trip.rename({"cls_attr" : "rating",
                   "name" : "user_name",
                   "food" : "menu"}, axis="columns")
```

```python
trip.head(1)
```

```python
trip = trip[["user_name", "res_name", "rating", "rev_title", "review", "menu"]]
```
