```python
import random
import pandas as pd
```

```python
df= pd.read_csv("./menu_taste_df_final.csv")
```

```python
df
```

```python
df["per"] = None
```

```python
df
```

```python
tmp = df[["spicy", "sweet", "sour", "saulty", "mild", "normal", "for_vegitarian", "for_mania", "for_starter"]]
```

```python
tmp
```

```python
sum(tmp.iloc[0])
```

```python
27/sum(tmp.iloc[0]) * 100
```

```python
for i in range(len(df)):
    df["per"][i] = sum(tmp.iloc[i])
```

```python
df
```

```python
taste = ["spicy","sweet","sour","saulty","mild","normal","for_vegitarian","for_mania","for_starter"]
```

```python
for i,j in enumerate(df["per"]):
    for k in taste:
        df[k].iloc[i] = round(df[k][i]/j * 100, 2)
```

```python
df.to_csv("./taste_per.csv", index=False)
```
