### 데이터 형식 바꾸기

```python
# 한줄씩 row 데이터를 뽑아서
foods = list()
cols  = test.columns
for i in tqdm(range( test.shape[0])):
    row_tmp = list()
    tmp = test.iloc[i, :]
    for j in range( tmp.shape[0]):
        # NaN이 아닌 컬럼만 획득
#         print(tmp[j])
        if tmp[j] == 1.0:
            row_tmp.append( cols[j] )
        else:
            row_tmp.append( np.nan )
#         # 그 컬럼명만 리스트에 담는다         
    foods.append( row_tmp )    
```
