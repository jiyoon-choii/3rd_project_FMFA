## 별점으로 긍정부정 컬럼 만들기

```python
import pandas as pd
```

```python
trip = pd.read_csv("./data/trip_final.csv")
```

```python
trip = trip.drop(["Unnamed: 0"], axis=1)
```

```python
trip.head(3)
```

```python
trip["P/N"] = None
```

```python
trip.head(3)
```

```python
for i,j in enumerate(trip["rating"]):
    if j >= 4.0 :
        trip["P/N"][i] = 1
    else:
        trip["P/N"][i] = 0
```

```python
sentiment_2 = trip[["review","P/N"]]
```
