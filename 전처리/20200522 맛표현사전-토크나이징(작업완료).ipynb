- 1. df['review']Ïùò Í∞í(Î¨∏Ïû•)ÏùÑ Îã®Ïñ¥ Îã®ÏúÑÎ°ú ÎÇòÎàÑÍ∏∞ 

```python
tag_count = []
tags = []

for review in df['review'][:1000] :
    # review Ïùò ÏÉÅÏúÑ row 10000Í∞ú Í∞ÄÏ†∏Ïò§Í∏∞ 
    # review = review[i] 
    word_tokens = word_tokenize(review)
    # review Í∞Ä Î¨∏Ïû•Ïù¥Í∏∞ ÎïåÎ¨∏Ïóê, 
    # Î¨∏Ïû•ÏùÑ Îã®Ïñ¥ Îã®ÏúÑÎ°ú ÎÇòÎà†Ï§å
    count = Counter(word_tokens)
    # Í∞Å Îã®Ïñ¥Í∞Ä Î™á Í∞úÏù∏ÏßÄ count Ïóê Îã¥Í∏∞ 
    
    for n,c in count.most_common(100) :
        # count.mos_common(n) : Í∞ÄÏû• ÎπàÎèÑ ÏàòÍ∞Ä ÎÜíÏùÄ Îã®Ïñ¥ n Í∞úÎ•º ÎΩëÎäîÎã§ 
        # n : Îã®Ïñ¥ / c : ÎπàÎèÑ
#         print(c)
        dics = {'tag':n, 'count':c}
        tag_count.append(dics)
        # [{Îã®Ïñ¥,ÎπàÎèÑ},{},....] Î•º tag_count Î¶¨Ïä§Ìä∏ Ïóê Îã¥ÏïÑÏ§ÄÎã§
        tags.append(dics['tag'])
        # [dicsÏùò value(c),c,c,......] Î•º tags Î¶¨Ïä§Ìä∏Ïóê Îã¥ÏïÑÏ§ÄÎã§ 
        
```
