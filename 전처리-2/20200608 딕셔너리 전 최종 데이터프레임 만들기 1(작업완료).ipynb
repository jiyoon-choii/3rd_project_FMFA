```python
# 오타 잡는 함수
def che(str1, str2):    
    st1 = list(0 for i in range(27))
    st2 = list(0 for j in range(27))
    str1=str1.strip()
    str2=str2.strip()
    if len(st1)>=len(st2):
        ss=len(st1)
    else:
        ss=len(st2)
    # print(ss)
    for i in str1:
        if ord(i)==32:
            st1[-1]+=1
        else:
            st1[ord(i)-97]+=1

    # print(st1)
    for i in str2:
        if ord(i)==32:
            st2[-1]+=1
        else:
            st2[ord(i)-97]+=1
    # print(st2)
    cnt = 0
    le=0
    for i in range(len(st1)):
        if st1[i]==st2[i]:
            cnt+=1
    if (cnt/ss) >=0.95:
        # print(cnt/ss)
        return 1
    else:
        # print(cnt/ss)
        return 0
```
